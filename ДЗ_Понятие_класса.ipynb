**Задание 1**
Напишите функцию, которая возвращает название валюты (поле ‘Name’) с максимальным значением курса с помощью сервиса www.cbr-xml-daily.ru...ly_json.js

```python
import requests
```

```python
currency = requests.get('https://www.cbr-xml-daily.ru/daily_json.js')

def exchange(currency):
  line = currency.json()['Valute']
  for key, value in line.items():
    currency_name = dict(value)
    if currency_name['Value'] > currency_name['Previous']:
      print(currency_name['Name'],currency_name['Value'])
    else:
      print(currency_name['Name'],currency_name['Previous'])
    
exchange(currency)
```
